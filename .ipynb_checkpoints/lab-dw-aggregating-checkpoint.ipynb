1. Create a new DataFrame that only includes customers who have a total_claim_amount greater than $1,000 and have a response of "Yes" to the last marketing campaign.

2. Using the original Dataframe, analyze the average total_claim_amount by each policy type and gender for customers who have responded "Yes" to the last marketing campaign. Write your conclusions.

4. Find the maximum, minimum, and median customer lifetime value by education level and gender. Write your conclusions.

```python
import pandas as pd
url = 'https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis.csv'
df = pd.read_csv(url)
df.columns
```

```python
filtered_df = df[(df["Total Claim Amount"]>1000) & (df["Response"]=="Yes")]
filtered_df.head()
```

```python
new_df = df[df["Response"]=="Yes"]
new_df.groupby(["Gender", "Policy Type"]).agg({"Total Claim Amount":"mean"})
```

```python
## Conclusion:
# Women generally have on average a bit higher claim amount, while for male special auto is lowest, for female its
# the highest. all values lie relativelely close to each other and around 1000 -1200
```

```python
df_states = df["State"].value_counts().reset_index()
df_states = df_states[df_states["State"]>500]
df_states
```

```python
df.groupby(["Education", "Gender"]).agg({"Customer Lifetime Value":["max", "min", "mean"]})
```

```python
## conclusion:
# no real correllation between level of education and Customer Lifetime Value. Doctor quite low in Comparision, maybe
# due to already old age. min for Doctor and Master higher then the rest. Maximum higher for younger people(College, high
# high school, bachelor). Even though max for Master is 2nd lowest, mean quite high. Mean mostly better for woman or
# equal.
```
